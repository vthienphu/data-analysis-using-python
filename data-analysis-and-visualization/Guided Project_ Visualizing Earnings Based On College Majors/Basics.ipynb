#### Import pandas and matplotlib into the environment.
#### Run the Jupyter magic %matplotlib inline so that plots are displayed inline.

```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# Read recent-grads.csv into pandas 
# and assign the resulting DataFrame to recent_grads.
recent_grads = pd.read_csv('recent-grads.csv')
```

```python
# return the first row formatted as a table.
recent_grads.iloc[0]
```

```python
# how the data is structured.
recent_grads.head()
```

```python
# how the data is structured.
recent_grads.tail()
```

```python
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print(raw_data_count)
print(cleaned_data_count)
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size', figsize=(10,5))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size', figsize=(10,5))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full_time', figsize=(10,5))
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. ShareWomen', figsize=(10,5))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men', figsize=(10,5))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Women', figsize=(10,5))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(7,15))
for index in range(len(cols)):
    ax = fig.add_subplot(len(cols),1,index+1)
    ax = recent_grads[cols[index]].plot(kind='hist', bins=25)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(6,6))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```
